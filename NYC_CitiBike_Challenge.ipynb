```python
import pandas as pd
import os
```

```python
# load Citi bike data
bikeshare_data_to_load = os.path.join("", "201908-citibike-tripdata.csv")
```

```python
# 1. Create a DataFrame for the 201908-citibike-tripdata data. 
bikeshare_data_df = pd.read_csv(bikeshare_data_to_load)
bikeshare_data_df.head()
```

```python
# 2. Check the datatypes of your columns. 
bikeshare_data_df.dtypes
```

```python
# 3. Convert the 'tripduration' column to datetime datatype.
bikeshare_data_df["tripduration"] = pd.to_datetime(bikeshare_data_df["tripduration"], unit='s')
bikeshare_data_df
```

```python
# 4. Check the datatypes of your columns. 
bikeshare_data_df.dtypes
```

```python
# 5. Export the Dataframe as a new CSV file without the index.
bikeshare_data_df.to_csv('bikesharing_clean.csv', index=False)
```
